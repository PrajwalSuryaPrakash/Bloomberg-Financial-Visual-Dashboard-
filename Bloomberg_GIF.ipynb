```python


# In[12]:


# Set a non-interactive backend for matplotlib
import matplotlib
matplotlib.use('Agg')

end = pd.Timestamp.now()
start = end - pd.DateOffset(days=120)

# Define the ticker and the period
ticker1 = 'TSLA US Equity'  # Tesla Inc
period = '10y'   

ticker2 = 'AAPL'

# Fetch data from Yahoo Finance
#data = yf.download(ticker1, period=period, interval='1d')  # Use daily data
#data1 = yf.download(ticker2, period=period, interval='1d')  # Use daily data

data = blp.bdh(tickers=ticker1, flds=['PX_LAST'],
                    start_date=start, end_date=end,
                    Per='D', Fill='P', Days='A', adjust='all')

print(data.index)
print(data)
```

```python
# Check if data is available
if data is None or data.empty:
    raise ValueError("No data fetched from Yahoo Finance. Please check the ticker symbol and period.")

# Print the first few rows of data to verify
print(data.head())
#print(data1.head())

```

```python
# Extract dates and closing prices
dates1 = data.index
#dates2 = data1.index
prices1 = data['TSLA US Equity', 'PX_LAST']
#prices2 = data1['Close']

print(prices1)
```

```python
df = pd.DataFrame(prices1)
```

```python
prices1 = list(prices1.reset_index(drop = True).values)
```

```python
# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Function to update the plot
def update(frame):
    if frame > len(dates1):
        return
    ax.clear()  # Clear previous content
    
    print(f"Plotting frame {frame} with data length: {len(dates1[:frame])}")  # Debug print
    ax.plot(dates1[:frame], prices1[:frame], color='blue')  # Plot up to the current frame
    ax.set_title(f'{ticker1} Stock Price Over Time')
    ax.set_xlabel('Date')
    ax.set_ylabel('Price (USD)')
    
    # Set x and y limits
    ax.set_xlim([dates1.min(), dates1.max()])
    ax.set_ylim([min(prices1), max(prices1)])
    
    ax.grid(True)

    # Save the current frame
    file_name = f'{frame_dir}/frame_{frame:03d}.png'
    plt.savefig(file_name, bbox_inches='tight')  # Save with tight bounding box
    print(f"Saved frame {frame} as {file_name}")  # Debug print

# Create directory to store frames
frame_dir = r'C:\Users\navab\OneDrive\Desktop\Bhargav\project\demo'
if not os.path.exists(frame_dir):
    os.makedirs(frame_dir)
    print(f"Created directory: {frame_dir}")
else:
    print(f"Directory already exists: {frame_dir}")

# Save each frame as an image
for i in range(1, len(dates1) + 1):  # Ensure we include the last frame
    update(i)  # Update the plot for the frame

# Verify if frames are created
frame_files = [f'{frame_dir}/frame_{i:03d}.png' for i in range(1, len(dates1) + 1)]
print(f"Frame files to be used: {frame_files}")

# Check if frames actually exist
for frame_file in frame_files:
    if not os.path.exists(frame_file):
        print(f"Frame file does not exist: {frame_file}")

# Load all frame images and save as a GIF in the specified directory
gif_path = r'C:\Users\navab\OneDrive\Desktop\Bhargav\project\demo\stock_prices.gif'

try:
    images = [Image.open(frame_file) for frame_file in frame_files if os.path.exists(frame_file)]
    if images:
        images[0].save(gif_path, save_all=True, append_images=images[1:], duration=100, loop=0)
        print(f"GIF saved to: {gif_path}")
    else:
        print("No images were loaded. Please check the frame files.")
except Exception as e:
    print(f"Error while creating GIF: {e}")

# Clean up temporary frame files
for frame_file in frame_files:
    if os.path.exists(frame_file):
        os.remove(frame_file)
```
